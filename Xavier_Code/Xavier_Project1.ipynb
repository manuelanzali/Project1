```python
#import 
import csv
import os
import pandas as pd
from pathlib import Path 
import matplotlib.pyplot as plt
```

```python
#Read in CSV file

csvpath = Path('Time_Wasters.csv')

# CSV reader specifies delimiter and variable that holds contents
df = pd.read_csv(csvpath, delimiter=',')

df.head()




```

```python
#Create datatframe for platforms

platforms_df = pd.DataFrame(df.groupby('Platform')['Total Time Spent'] .mean())
platforms_df

```

```python
#Create datatframe for Total Time

total_time_df = pd.DataFrame(df.groupby(['Platform', 'Video Category'])['Total Time Spent'].mean())
total_time_df
```

```python
#Create datatframe for Satisfaction

satisfaction_df = pd.DataFrame(df.groupby(['Platform', 'Video Category'])['Satisfaction'].mean())
satisfaction_df

```

```python
#Identify my Variables
platform_types = platforms_df.index
column_values = platforms_df['Total Time Spent']

plt.bar(platform_types, column_values)
plt.xlabel('Platform')
plt.ylabel('Total Time Spent')
plt.xticks(rotation=45)
plt.title('Average Time Spent on Social Media')
plt.show()
```
