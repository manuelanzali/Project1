```python
import pandas as pd
from pathlib import Path
```

```python
# File to Load (Remember to Change These)
socialmedia_data_to_load = Path("Time-Wasters on Social Media.csv")

# Read Social media Data File and store into Pandas DataFrames
socialmedia_data = pd.read_csv(socialmedia_data_to_load)

socialmedia_data
```

```python
#Correct a misspelled country and lowercase profession in the DataFrame
socialmedia_data["Location"] = socialmedia_data["Location"].replace("Barzil", "Brazil")
socialmedia_data["Profession"] = socialmedia_data["Profession"].replace("driver", "Driver")
socialmedia_data
```

```python
# Calculate the total number of unique professions
profession_count = len(socialmedia_data["Profession"].unique())
profession_count
```

```python
# Calculate the total number of Users
User_count = socialmedia_data["UserID"].count()
User_count
```

```python
#Clean up and rename columns in DF
df = socialmedia_data.rename(columns={"Location" : "Country"})
df.head()
```

```python
# Calculate the total number of unique locations
location_count = len(df["Country"].unique())
location_count
```

```python
# Group data by location
location_grouped = df.groupby("Country")
print(location_grouped)

#Print number of users in each location
for Country, group_data in location_grouped:
    print(f"Country: {Country}")
    user_counts = group_data["UserID"].count()
    print(user_counts)
```

```python
#Calculate the average productive time spent in each country
average_productive_time_per_country = df.groupby("Country")["ProductivityLoss"].mean()
print(average_productive_time_per_country)
```

```python
#platform - age satisfactions?
# find the max and min of age
max_age = clean_df["Age"].max()
min_age = clean_df["Age"].min()
print(min_age)
print(max_age)


```

```python
#bin
#When looking at each country and platform and age group, show satisfaction in each group 
#average productivity loss by bin

# Establish the bins.
age_bins = [0, 21, 42, 64]
labels = ["Youth (<21)", "Adult(21-42)", "Older Adult (42-64)"]

# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

clean_df["Age Group"] = pd.cut(clean_df["Age"], 
                                        age_bins, labels=labels, include_lowest=True)


clean_df
```
